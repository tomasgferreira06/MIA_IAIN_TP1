# Genetic Algorithm for Job Scheduling Problem

## Overview
The **Job Scheduling Problem (JSP)** is a classic optimization problem where the objective is to allocate a set of $N$ jobs to $M$ machines, optimizing a specific performance criterion such as minimizing the **makespan** (the total time to complete all jobs). Each job consists of a sequence of operations that must be performed in a specific order, and each operation must be processed on a specific machine. The scheduling process must respect constraints like machine availability and the sequence dependency of operations.

---

## Key Components of the Problem

### 1. Jobs and Operations
- A job is a task or set of tasks (operations) that must be completed.
- Each job has a fixed order of operations. For example, Job $J_1$ might require Machine $M_1$ for the first operation, $M_2$ for the second, and so on.
- The processing time of each operation is predetermined and specific to the job-machine pair.

### 2. Machines
- There are $M$ machines available for processing jobs.
- At any given time, a machine can process only one job, and an operation cannot begin until the required machine is available.

### 3. Processing Times
- Each operation has a fixed processing time on the respective machine. These times are represented as a matrix:
  - Rows correspond to jobs.
  - Columns correspond to machines.
  - The element at $(i, j)$ is the processing time of Job $i$ on Machine $j$.

### 4. Constraints
- **Sequence Dependency:** Operations of a job must be processed in the given order.
- **Machine Availability:** A machine can only process one operation at a time.
- **No Preemption:** Once an operation starts, it must run to completion without interruption.

### 5. Objective
- The primary objective is to minimize the **makespan**, which is the total time required to complete all jobs.
- Secondary objectives could include minimizing idle time for machines or balancing workloads across machines.

---

**Goal:** Schedule the jobs on the machines to minimize the makespan.

---

## Challenges in JSP
1. **Combinatorial Complexity:**
   - As the number of jobs ($N$) and machines ($M$) increases, the number of possible schedules grows exponentially.
   
2. **Conflict Resolution:**
   - Jobs competing for the same machine at the same time create conflicts that need to be resolved optimally.

3. **Dependency Constraints:**
   - The precedence order of operations within each job must be maintained.

4. **Balancing Trade-offs:**
   - Trade-offs between optimizing for makespan, machine utilization, or job prioritization must be addressed.

---

## Applications of JSP
The Job Scheduling Problem has numerous real-world applications, including:
1. **Manufacturing Systems:** Scheduling tasks in factories with multiple workstations.
2. **Cloud Computing:** Allocating virtual machines to tasks in a data center.
3. **Healthcare:** Scheduling surgeries or diagnostic procedures across limited resources.
4. **Logistics and Transportation:** Allocating vehicles or delivery routes to shipments.

## Problem Setup

```python
# Problem Parameters
NUM_JOBS = 20 # N jobs
NUM_MACHINES = 5 # M machines
POP_SIZE = 200
NUM_GENERATIONS = 100
CROSSOVER_RATE = 0.6
MUTATION_RATE = 0.05

# Randomly generated processing times for jobs and machines 
np.random.seed(200)

processing_times = np.random.randint(1, 20, size=(NUM_JOBS, NUM_MACHINES))
print("Processing Times:\n", processing_times)
```

## Evaluation Function

The solutions are evaluated by calculating the following function. The evaluation function assesses the quality of a given job schedule by calculating its makespan—**the total time required to complete all jobs across multiple machines**. It does this by simulating the processing of each job in the specified order on each machine, ensuring that each job starts only after both the machine is available and the job has finished processing on the previous machine. By tracking the finish times of both machines and jobs, the function determines the maximum completion time across all machines, which represents the makespan. In the context of optimization or genetic algorithms, this makespan serves as the fitness score, where lower values indicate more efficient schedules. Thus, the fitness function guides the search for optimal job sequences that minimize the overall completion time.

```python
# Objective: Minimize makespan
def evaluate_schedule(schedule, processing_times):
    job_sequence = schedule
    num_jobs, num_machines = processing_times.shape
    machine_times = np.zeros(num_machines)
    job_times = np.zeros(num_jobs)

    for job in job_sequence:
        for machine in range(num_machines):
            start_time = max(machine_times[machine], job_times[job])
            end_time = start_time + processing_times[job][machine]
            machine_times[machine] = end_time
            job_times[job] = end_time
            
    return max(machine_times)

resultado = evaluate_schedule([0,1,2],np.array([[5,8],[3,6],[4,7]]))
print(resultado)
```

## Parent Selection

Selecting parents is an essential step in Evolutionary Algorithms, as we aim to encourage the propagation of interesting traits over the generations. Complete the cells below to implement the tournament and roulette wheel selection methods.

```python
def tournament_selection(tourn_size):
    def select(population):
        """
        Seleciona Um indivíduo da população usando o torneio.

        1. Escolhe 'tourn_size' indivíduos aleatórios da população (sem repetir).
        2. Entre esses, escolhe o que tem menor fitness (melhor solução).
        3. Devolve esse indivíduo.
        """
        # 1. escolher 'tourn_size' indivíduos aleatórios da população
        contenders = random.sample(population, tourn_size)

        # 2. escolher o melhor entre esses contenders (menor fitness)
        best = min(contenders, key=lambda ind: ind['fitness'])

        # 3. devolver esse indivíduo
        return best
    # devolve a função select, já configurada com este tourn_size
    return select
```

```python
def roulette_wheel_selection(population):
    """
    Seleciona um indivíduo da população usando Roulette Wheel Selection.

    Como o fitness é para minimizar (makespan), convertemos fitness em aptidão:
        aptidão = 1 / (fitness + ε)

    Quanto menor o fitness → maior a aptidão → maior probabilidade.
    """

    eps = 1e-6  # para evitar divisão por zero

    # 1. Ir buscar os fitness de cada indivíduo
    fitness_values = [ind['fitness'] for ind in population]

    # 2. Converter fitness em aptidão inversa
    fitness_inv = [1.0 / (f + eps) for f in fitness_values]

    # 3. Calcular soma total das aptidões
    total = sum(fitness_inv)

    # 4. Gerar número aleatório entre 0 e total
    r = random.random() * total

    # 5. Somar aptidões até passar r e selecionar o indivíduo
    acc = 0.0
    for ind, weight in zip(population, fitness_inv):
        acc += weight
        if acc >= r:
            return ind

    # Caso algum erro numérico falhe a seleção:
    return population[-1]
```

## Survivor Selection

The survivor selection mechanism is responsible for creating a new population based on the available individuals. Various techniques do exist. Complete the cell below to implement the elitist selection mechanism.

```python
def elitist_selection(elite_size):
    def select(population, offspring):
        """
        Seleção elitista de sobreviventes.

        - Mantém 'elite_size' melhores indivíduos da população antiga (pais).
        - Preenche o resto da nova população com os melhores filhos (offspring),
          até ficar com exatamente POP_SIZE indivíduos.
        """

        # 1. Ordenar a população antiga por fitness (menor = melhor)
        sorted_pop = sorted(population, key=lambda ind: ind['fitness'])
        elite = sorted_pop[:elite_size]  # top 'elite_size'

        # 2. Ordenar os filhos por fitness
        sorted_offspring = sorted(offspring, key=lambda ind: ind['fitness'])

        # 3. Quantos filhos precisamos para completar a nova população?
        num_offspring_needed = POP_SIZE - elite_size

        # 4. Escolher os melhores filhos suficientes
        best_offspring = sorted_offspring[:num_offspring_needed]

        # 5. Nova população = elite (pais) + melhores filhos
        new_pop = elite + best_offspring
        
        return new_pop
    return select
```

## Genetic Algorithm for Binary Representation


The code below uses a binary representation of the individuals to implements a **priority** for job scheduling. Each job is assigned a binary value (0 or 1) that serves as its primary priority level. Jobs with priority 0 are considered "high priority", while those with priority 1 are considered "low priority". 

Prior to being evaluated, each individual's genotype must be mapped to its phenotype, yielding a permutation of jobs. Consider the example below:

The binary genotype `[0, 1, 0, 1]` for 4 jobs translates to:
- **High Priority (0):** Jobs 0, 2 are scheduled first in order [0, 2]
- **Low Priority (1):** Jobs 1, 3 are scheduled second in order [1, 3] 
- **Final Schedule:** [0, 2, 1, 3]



### Crossover

The crossover operator is responsible for creating new individuals by combining the genes of the parents. Complete the cells below to implement an one-point crossover and an uniform crossover.

```python
def one_point_crossover(parent1, parent2):
    """
    Crossover de um ponto para representação binária.

    - parent1 e parent2 são listas de 0s e 1s (tamanho = NUM_JOBS).
    - Escolhemos um índice de corte (point).
    - child1: início de parent1 + fim de parent2.
    - child2: início de parent2 + fim de parent1.
    """
    length = len(parent1)

    # Escolher um ponto de corte entre 1 e length-1
    point = random.randint(1, length - 1)

    # Construir os dois filhos
    child1 = parent1[:point] + parent2[point:]
    child2 = parent2[:point] + parent1[point:]
    return child1, child2
```

```python
def uniform_crossover(parent1, parent2):
    """
    Uniform crossover para representação binária.

    Para cada gene:
    - Com 50% de probabilidade, o gene de child1 vem de parent1 e child2 de parent2.
    - Com 50% de probabilidade, trocam-se os papéis.
    """

    length = len(parent1)
    child1 = []
    child2 = []

    for i in range(length):
        # Sorteia se este gene vem de parent1 ou parent2
        if random.random() < 0.5:
            # gene não trocado
            child1.append(parent1[i])
            child2.append(parent2[i])
        else:
            # trocados
            child1.append(parent2[i])
            child2.append(parent1[i])
    return child1, child2
```

### Mutation

The mutation operator is responsible for introducing new genetic material into the population, by making small changes to an individual's genotype. Complete the cell below to implement the flip mutation.

```python
def bit_flip_mutation(individual):
    """
    Aplica mutação bit flip:
    - Para cada gene (bit) do indivíduo,
    - Com probabilidade MUTATION_RATE,
      o bit é invertido (0→1, 1→0).
    """
    for i in range(len(individual)):
        if random.random() < MUTATION_RATE:
            individual[i] = 1 - individual[i]  # flip do bit
```

### Evolutionary cycle

Using the mechanisms defined in the previous cells, configure and run the binary_ga from the cell below. 

```python
def binary_ga(parent_selection, crossover, mutation, survivor_selection):
    # Initialize Population
    population = [[random.randint(0, 1) for _ in range(NUM_JOBS)] for _ in range(POP_SIZE)]
    decoded_population = [sorted(range(NUM_JOBS), key=lambda x: ind[x]) for ind in population]     
    fitness = [evaluate_schedule(ind, processing_times) for ind in decoded_population]
    population = [{'genotype': population[i], 'phenotype': decoded_population[i], 'fitness': fitness[i]} for i in range(POP_SIZE)]
    population.sort(key=lambda x: x['fitness'], reverse=False)
    print('Best of the initial population:', population[0])
    
    best_fitness_progress = [population[0]['fitness']]
    for gen in range(NUM_GENERATIONS):
        # Parent Selection
        parents = [parent_selection(population) for _ in range(POP_SIZE)]
        
        # Reproduction
        offspring = []
        i = 0
        while(i<len(parents)-1):
            # Crossover
            if random.random()<CROSSOVER_RATE:
                o1, o2 = crossover(parents[i]['genotype'], parents[i+1]['genotype'])
            else:
                o1 = copy.deepcopy(parents[i]['genotype'])
                o2 = copy.deepcopy(parents[i+1]['genotype'])

            # Mutation
            mutation(o1)
            mutation(o2)
            offspring.append(o1)
            offspring.append(o2)
            i += 2
        
        # Evaluation
        decoded_offspring = [sorted(range(NUM_JOBS), key=lambda x: ind[x]) for ind in offspring]
        fitness = [evaluate_schedule(ind, processing_times) for ind in decoded_offspring]
        offspring = [{'genotype': offspring[i], 'phenotype': decoded_offspring[i], 'fitness': fitness[i]} for i in range(POP_SIZE)]
        offspring.sort(key=lambda x: x['fitness'], reverse=False)
        
        # Survivor Selection
        population = survivor_selection(population, offspring)
        population.sort(key=lambda x: x['fitness'], reverse=False)
        best_fitness_progress.append(population[0]['fitness'])
        
        print('Best of generation',gen,'fitness: ',population[0]['fitness'], 'phenotype: ',population[0]['phenotype'] )

    return population[0]['phenotype'], population[0]['fitness'], best_fitness_progress
```

## Genetic Algorithm for Permutation Representation

The permutation representation uses a direct encoding where each individual represents an explicit job ordering without requiring any decoding step. Consider the following example, where an individual's genotype contains the information `[2, 0, 3, 1]`:
- This directly represents the job processing order: Job 2 → Job 0 → Job 3 → Job 1
- No conversion between genotype and phenotype is needed



### Crossover

The crossover operator is responsible for creating new individuals by combining the genes of the parents. Complete the cells below to implement an one-point crossover and an uniform crossover.

```python
def order_crossover(parent1, parent2):
    """
    Implementação do Order Crossover (OX) para representação por permutação.
    Garante que não há duplicados e que a ordem relativa dos genes é preservada.
    """

    length = len(parent1)

    # 1. Escolher dois pontos de corte
    start = random.randint(0, length - 2)
    end = random.randint(start + 1, length - 1)

    # 2. Criar filhos vazios
    child1 = [None] * length
    child2 = [None] * length

    # 3. Copiar o segmento de parent1 → child1
    child1[start:end+1] = parent1[start:end+1]
    #    Copiar o segmento de parent2 → child2
    child2[start:end+1] = parent2[start:end+1]

    # 4. Preencher o resto com a ordem do outro pai
    def fill_child(child, parent_other):
        pos = (end + 1) % length
        for gene in parent_other:
            if gene not in child:
                child[pos] = gene
                pos = (pos + 1) % length

    fill_child(child1, parent2)
    fill_child(child2, parent1)

    return child1, child2
```

### Mutation

The mutation operator is responsible for introducing new genetic material into the population, by making small changes to an individual's genotype. Complete the cell below to implement the flip mutation.

```python
def swap_mutation(individual):
    """
    Mutação por troca (swap):
    - Com probabilidade MUTATION_RATE,
      escolhe duas posições aleatórias do cromossoma
      e troca os valores entre elas.
    """
    if random.random() < MUTATION_RATE:
        length = len(individual)

        # escolher duas posições diferentes
        i, j = random.sample(range(length), 2)

        # trocar os elementos
        individual[i], individual[j] = individual[j], individual[i]
```

```python
def scramble_mutation(individual):
    """
    Scramble Mutation:
    - Seleciona um segmento aleatório da permutação.
    - Baralha os elementos dentro desse segmento.
    - Mantém o resto do indivíduo igual.
    """
    if random.random() < MUTATION_RATE:
        length = len(individual)

        # 1. escolher segmento
        start = random.randint(0, length - 2)
        end = random.randint(start + 1, length - 1)

        # 2. extrair segmento
        subset = individual[start:end+1]

        # 3. baralhar
        random.shuffle(subset)

        # 4. reinserir no indivíduo
        individual[start:end+1] = subset
```

### Evolutionary cycle

Using the mechanisms defined in the previous cells, configure and run the permutation_ga from the cell below. 

```python
def permutation_ga(parent_selection, crossover, mutation, survivor_selection):
    # Initialize Population
    population = [random.sample(range(NUM_JOBS), NUM_JOBS) for _ in range(POP_SIZE)]
    # Evaluate Initial Population
    fitness = [evaluate_schedule(ind, processing_times) for ind in population]
    
    # Convert to dictionary format
    population = [{'genotype': population[i], 'fitness': fitness[i]} for i in range(POP_SIZE)]

    # Sort population by fitness (note that this is a minimization problem)
    population.sort(key=lambda x: x['fitness'], reverse=False)
    print('Best of the initial population:', population[0])

    best_fitness_progress = [population[0]['fitness']]
    for gen in range(NUM_GENERATIONS):
        # Parent Selection    
        parents = [parent_selection(population) for _ in range(POP_SIZE)]        

        # Reproduction
        offspring = []
        i = 0
        while(i<len(parents)-1):
            # Crossover
            if random.random()<CROSSOVER_RATE:
                o1, o2 = crossover(parents[i]['genotype'], parents[i+1]['genotype'])
            else:
                o1 = copy.deepcopy(parents[i]['genotype'])
                o2 = copy.deepcopy(parents[i+1]['genotype'])

            # Mutation
            mutation(o1)
            mutation(o2)
            offspring.append(o1)
            offspring.append(o2)
            i += 2
            
        # Evaluation
        fitness = [evaluate_schedule(ind, processing_times) for ind in offspring]
        offspring = [{'genotype': offspring[i], 'fitness': fitness[i]} for i in range(POP_SIZE)]
        offspring.sort(key=lambda x: x['fitness'], reverse=False)
        
        # Survivor Selection
        population = survivor_selection(population, offspring)
        population.sort(key=lambda x: x['fitness'], reverse=False)
        
        best_fitness_progress.append(population[0]['fitness'])
        
        print('Best of generation',gen,'fitness: ',population[0]['fitness'], 'genotype: ',population[0]['genotype'] )

    return population[0]['genotype'], population[0]['fitness'], best_fitness_progress
```
